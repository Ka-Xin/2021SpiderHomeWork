```python
import requests
from lxml import etree
```

```python
list_name = ["sjbz", "padbz", "bjbbz", "dnbz", "dsbz"]
headers={
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36'
    }
```

```python
for name in list_name:
    for num in range(1,6):
        url = "http://www.douikan.com/" + name + "/p" + str(num) +"/"
        response = requests.get(url=url, headers=headers)
        html = etree.HTML(response.text)
        pic_list = html.xpath('//div[@class="all_list"]//a/img')
        for i in range(len(pic_list)):
            pic_url = pic_list[i].xpath("./@src")
            pic_name = pic_list[i].xpath("./@alt")
            pic_response = requests.get(url=pic_url[0], headers=headers)
            pic_name = pic_name[0].replace('?',"-")
            with open('06_pic/' + pic_name + '.jpg', 'wb') as fp:
                fp.write(pic_response.content)
                print(pic_name+"完成！")
```
