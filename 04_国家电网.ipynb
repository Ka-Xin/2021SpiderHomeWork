```python
import requests
import json

url_id = "https://ecp.sgcc.com.cn/ecp2.0/ecpwcmcore//index/noteList"
headers={
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36'
    }
f = open("电网招标.txt","w+",encoding="utf-8")
page = 0
for i in range(1,6):
    page+=1
    data = {
        'firstPageMenuId': "2018032700291334",
        'index': i,
        'key': "",
        'orgId': "",
        'orgName': "",
        'purOrgCode': "",
        'purOrgStatus': "",
        'purType': "",
        'size': 20
    }
    response = requests.post(url=url_id,headers=headers,data=json.dumps(data))
    js = json.loads(response.text)
    list = []
    item=0
    for i in range(20):
        ID = js["resultValue"]["noteList"][i]["id"]
        list.append(ID)
    for i in range(20):
        try:
            url_content = "https://ecp.sgcc.com.cn/ecp2.0/ecpwcmcore//index/getNoticeBid"
            data = list[i]
            response = requests.post(url=url_content,headers=headers,json=data)
            js = json.loads(response.text)
            list_js = js["resultValue"]["notice"]
            content = ("采购项目名称:",list_js["PURPRJ_NAME"],
                "采购项目编号:",list_js["PURPRJ_CODE"],
                "采购类型:",list_js["PUR_TYPE_NAME"],
                "招标人:",list_js["BID_ORG"],
                "联系人:",list_js["CONTACT"],
                "联系电话:",list_js["TEL"],
                "电子邮箱:",list_js["E_MAIL"],
                "开标（截标）时间:",list_js["OPENBID_TIME"],
                "招标文件获取截止时间:",list_js["BIDBOOK_BUY_END_TIME"]  
            )
        except:
            try:
                url_content = "https://ecp.sgcc.com.cn/ecp2.0/ecpwcmcore//index/getChangeBid"
                data = list[i]
                response = requests.post(url=url_content,headers=headers,json=data)
                js = json.loads(response.text)
                list_js = js["resultValue"]["notice"]
                content = ("采购项目名称:",js["resultValue"]["origNotice"]["CHG_NOTICE_CONT"],
                    "采购项目编号:",js["resultValue"]["origNotice"]["PURPRJ_CODE"],
                    "采购类型:",js["resultValue"]["origNotice"]["PUR_TYPE_NAME"],
                    "招标人:",js["resultValue"]["origNotice"]["BID_ORG"],
                    "联系人:",js["resultValue"]["origNotice"]["CONTACT"],
                    "联系电话:",js["resultValue"]["origNotice"]["TEL"],
                    "电子邮箱:",js["resultValue"]["origNotice"]["E_MAIL"],
                    "开标（截标）时间:",js["resultValue"]["origNotice"]["OPENBID_TIME"],
                    "招标文件获取截止时间:",js["resultValue"]["origNotice"]["BIDBOOK_BUY_END_TIME"])
            except:
                    url_content = "https://ecp.sgcc.com.cn/ecp2.0/ecpwcmcore//index/getDoc"
                    data = list[i]
                    response = requests.post(url=url_content,headers=headers,json=data)
                    js = json.loads(response.text)
                    list_js = js["resultValue"]["doc"]
                    content = ("采购项目名称:",list_js["purOrgName"],
                        "采购项目编号:",list_js["purOrgCode"],
                        "采购类型:",list_js["purName"],
                        "招标人:",list_js["publishOrgName"])
        f.write(str(content))
        item+=1
        print("完成第",page,"页","第",item,"项")
```
